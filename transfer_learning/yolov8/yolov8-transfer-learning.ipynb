# Kiểm tra số file

```python
import os

def find_missing_folders(folder_path, total_folders):
    existing_folders = [int(f) for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
    missing_folders = [f for f in range(total_folders) if f not in existing_folders]
    return missing_folders

train_folder =  "/content/gdrive/My Drive/YOLOv8_sinonom/dataset/wb_recognition_dataset/train"
total_folders = 2139  # Số lượng thư mục con cần kiểm tra

missing_folders = find_missing_folders(train_folder, total_folders)

if missing_folders:
    print("Các số folder bị thiếu:")
    print(missing_folders)
else:
    print("Không có folder nào bị thiếu trong khoảng từ 0 đến 2138.")
```
